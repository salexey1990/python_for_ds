```python
def prepare_square(df):
    df.loc[df['Square']>(df['Rooms']*40+40),'Square']=df['Rooms']*40+40
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    return df

def prepare_lifesquare(df):
    df.loc[df['Square']<df['LifeSquare'],'LifeSquare']=df['Square']
    return df

def fillna_life_square(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    return df

def join_stats1(df,stats1,source_df,mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId','Rooms'],how='left')
    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
stats1=data.groupby(['DistrictId','Rooms'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
data=prepare_square(data)
data=prepare_lifesquare(data)
data=fillna_life_square(data,data)
data=join_stats1(data,stats1,data)
```

```python
data.info()
```

```python
data.describe()
```

```python
features=data[['Rooms','Square','LifeSquare','HouseYear','mean_price']]
target=data['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score as r2
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(features,target)
```

```python
model=RF(n_estimators=100,max_depth=10,min_samples_leaf=5,random_state=42)
```

```python
model.fit(X_train,y_train)
y_pred=model.predict(X_valid)
```

```python
r2(y_valid,y_pred)
```
