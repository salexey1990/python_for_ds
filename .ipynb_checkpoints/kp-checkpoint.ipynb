```python
def prepare_kitchen_square(df):
    df.loc[df['KitchenSquare'] > df['LifeSquare'], 'KitchenSquare'] = df['KitchenSquare'].mean()
    df.loc[df['KitchenSquare']  > 50, 'KitchenSquare'] = df['KitchenSquare'].mean()
    return df
```

```python
def prepare_life_square(df):
    df.loc[df['LifeSquare'] > df['Square'], 'LifeSquare'] = df['Square']
    return df
```

```python
def fillna_life_square(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    return df
```

```python
data = fillna_life_square(data,data)
data = prepare_kitchen_square(data)
data = prepare_life_square(data)
```

```python
data = pd.get_dummies(data)
data.info()
```

```python
price=data['Price'].mean()
price_mean_by_district=data.groupby(['DistrictId', 'Rooms'],as_index=False).agg({'Price':'mean'}).\
                    rename(columns={'Price':'mean_price_by_district'})
```

```python
def add_mean_price(df,price_mean_by_district,price):
    df=pd.merge(df,price_mean_by_district,on=['DistrictId', 'Rooms'],how='left')
    df['mean_price_by_district']=df['mean_price_by_district'].fillna(price)
    return df
```

```python
data = add_mean_price(data, price_mean_by_district, price)
data.head()
```

```python
data.drop('DistrictId', axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse,r2_score as r2
from sklearn.ensemble import RandomForestRegressor as RF
model = RandomForestRegressor(n_estimators=1000, random_state=42, max_depth=15, min_samples_leaf=3)
```

```python
X = data.loc[:, data.columns != 'Price']
y = data['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
```

```python
r2(y_valid, y_pred)
```

```python
test = pd.read_csv('/home/alex/Downloads/Kurs_project_task/test.csv')
test.info()
```

```python
test = fillna_life_square(test,data)
test = prepare_kitchen_square(test)
test = prepare_life_square(test)
test = add_mean_price(test, price_mean_by_district, price)
test = pd.get_dummies(test)
test.drop('Healthcare_1', axis=1, inplace=True)
test.drop('DistrictId', axis=1, inplace=True)
```
